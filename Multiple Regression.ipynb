```python
#importing relevant packages for statistical analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
#loading data for analysis
data = pd.read_csv('real_estate_price_size_year.csv')
```

```python
#checking data
data
```

```python
data.describe()
```

```python
#Defining IVs and DV variables for analysis
y = data['price']
x1 = data[['size', 'year']]
```

```python
#Exploring the data visually -> DV price, IV size
plt.scatter(data['size'], y)
plt.xlabel('size', fontsize = 25)
plt.ylabel('price', fontsize = 25)
plt.show()
```

```python
#Exploring the data visually -> DV price, IV size
plt.scatter(data['year'], y)
plt.xlabel('year', fontsize = 25)
plt.ylabel('price', fontsize = 25)
plt.show()
```

```python
#Code for the analysis, ordinary least squares 
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
#Print results
results.summary()
```
